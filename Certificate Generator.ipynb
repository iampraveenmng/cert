# Script to Generate Certificates in bulk

```python
from PIL import Image, ImageDraw, ImageFont
import pandas as pd

form = pd.read_excel("fb.xlsx")

name_list = form['Name'].to_list()
email_list = form['Email'].to_list()
dummy_list = ["Names Go here for certificated"]

for i in dummy_list:
    im = Image.open("cert.jpg")
    d = ImageDraw.Draw(im)
    location = (275, 1050)
    text_color = (0, 137, 209)
    font = ImageFont.truetype("AlexBrush-Regular.ttf", 250, encoding="unic")
    d.text(location, i, fill=text_color,font=font)
    im.save("certificate_"+i+".pdf")
```

```python
import smtplib
import ssl
from email.mime.text import MIMEText
from email.utils import formataddr
from email.mime.multipart import MIMEMultipart  # New line
from email.mime.base import MIMEBase  # New line
from email import encoders  # New line

sender_email = #Your Email goes here in quotes ""
sender_name = "Mursal Furqan"
password = #You Password Goes here

receiver_emails = email_list
receiver_names = name_list
# receiver_emails = ['mursalfurqan@gmail.com']
# receiver_names = ['Mursal Furqan']

for receiver_email, receiver_name in zip(receiver_emails, receiver_names):
        
        
        print("Sending to " + receiver_name)
        msg = MIMEMultipart()
        msg['Subject'] = 'Certificate | ' + receiver_name + ' | MS Cognitive Serice & Custom Vision'
        msg['From'] = formataddr((sender_name, sender_email))
        msg['To'] = formataddr((receiver_name, receiver_email))
        
        msg.attach(MIMEText("""<h2>Hello again, """ + receiver_name + """</h2>
                            <p>Please find the attachment to this email for additional resources of the webinar b Mr.
                            Sudereshwar Kodi. Powered by Microsoft Student Partners Society, 
                            in collaboration with IEEE Computer Society, Mehran UET, Jamshoro, live on MS Teams on 27th June, 2020.</p>
                            <br>
                            
                            <p>We have an exciting series of resourceful events lined up for the future.</p>
                            <br>
                            <p>For further updates,do follow us on our social medial handles</p>
                            <ul>
                                <li><a href = 'https://www.facebook.com/msp.jam/'>Facebook</a></li>
                                <li><a href = 'https://www.instagram.com/mspmuet/'>Instagram</a></li>
                            </ul>
                            <h4>Have a great day</h4>
                            """, 'html'))
#         filename = "certificate_" + receiver_name + ".pdf"
        filename = "Additional_Resources.rar"

        try:
            with open(filename, "rb") as attachment:
                            part = MIMEBase("application", "octet-stream")
                            part.set_payload(attachment.read())

            encoders.encode_base64(part)
            part.add_header(
                    "Content-Disposition",
                    f"attachment; filename= {filename}",
            )

            msg.attach(part)
        except Exception as e:
                print(f'Oh no! We didnt found the attachment!n{e}')
                break

        try:
                server = smtplib.SMTP('smtp.office365.com', 587)
                context = ssl.create_default_context()
                server.starttls(context=context)
                server.login(sender_email, password)
                server.sendmail(sender_email, receiver_email, msg.as_string())
                print('Email sent!')
        except Exception as e:
                print(f'Oh no! Something bad happened!n{e}')
                break
        finally:
                print('Closing the server...')
                server.quit()
```
